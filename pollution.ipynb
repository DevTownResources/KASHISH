```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_air = pd.read_csv("poll.csv")
df_air.head(10)
```

```python
sns.set
plt.figure(figsize=(45, 15), dpi = 500)
sns.barplot(x='Date', y = 'AQI Value', data = df_air)
```

```python
sns.set()
plt.figure(figsize = (15, 45) , dpi = 500)
sns.barplot(x = 'AQI Value', y = 'Country', data = df_air)
```
