```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 5, 3])
y = np.array([2, 10, 5])


plt.title("rahul vitals", loc = 'right')
plt.xlabel("sugar")
plt.ylabel("blood pressure")
plt.plot(x, y)
plt.grid(color = 'blue', ls = '--')
plt.show()
```
