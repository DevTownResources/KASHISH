```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_production = pd.read_csv("production.csv")
df_domestic_consumption = pd.read_csv("consumption.csv")
df_export = pd.read_csv("export.csv")
df_import = pd.read_csv("import.csv")
df_re_export = pd.read_csv("reexport.csv")
```

```python
plt.style.use('fivethirtyeight')

fig = plt.figure(figsize=(31, 22))

plt.barh(df_production["Country"], df_production["Total_production"])

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.xlabel('Total coffee production', fontsize=24) 
plt.ylabel('Country', fontsize=24)

current_values = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_values])
        
plt.title("Total coffee production crop year from 1990 to 2020", fontsize=32)
        
plt.show()
```
